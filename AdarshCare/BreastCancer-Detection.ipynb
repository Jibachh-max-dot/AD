```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cancer.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# We would drop "Unnamed: 32", "id" from the dataset, as they are not beneficial for our analysis
df.drop(['Unnamed: 32', 'id'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['diagnosis'].nunique()
```

### Hence, we have "diagnosis" as the target variable, which has two values, M(Malign) and B(Benign)
### After doing PCA(Principal Component Analysis), we find out that features like 'concave points_mean','area_mean','radius_mean','perimeter_mean','concavity_mean' are important for our prediction

```python
df = df[['concave points_mean','area_mean','radius_mean','perimeter_mean','concavity_mean', 'diagnosis']]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.head(20)
```

```python
encoder = LabelEncoder()
diagnosisEncoded = encoder.fit_transform(df['diagnosis'])
df['diagnosis'] = diagnosisEncoded
```

```python
df.head(20)
```

```python
X = df.drop(['diagnosis'], axis=1)
y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
model = RandomForestClassifier()
model.fit(X_train, y_train)
predictionsRF = model.predict(X_test)
```

```python
def calculateMetrics(model):
    print(classification_report(y_test, model))
    print('\n')
    print(confusion_matrix(y_test, model))
```

```python
calculateMetrics(predictionsRF)
```
